```python
thetas = term_1.dot(term_2)
```

```python
thetas
```

```python
bias = thetas[0]
weight_1 = thetas[1]
weight_2 = thetas[2]
weight_3 = thetas[3]
```

```python
def predict(bias, weight_1, weight_2, weight_3, x_1, x_2, x_3):
    predicted_value = bias + (x_1 * weight_1) + (x_2 * weight_2) + (x_3 * weight_3)
    return predicted_value
```

```python
x_1 = float(input('enter the spend on YouTupe: '))
x_2 = float(input('enter the spend on FaceBook: '))
x_3 = float(input('enter the spend on NewsPaper: '))
```

```python
print('Predicted Sales is: {0}'.format(predict(bias, weight_1, weight_2, weight_3, x_1, x_2, x_3)))
```

```python
X['predicted_sales'] = X.apply(lambda row: predict(bias, weight_1, weight_2, weight_3, row['youtube'], row['facebook'], row['newspaper']), axis=1)
```

```python
data = pd.concat([data, X['predicted_sales']], axis=1)
```
